## **10.6. Múltiples Gráficos**

Para agregar múltiples gráficos al entorno figura, deben modificarse los parámetros del método `subplots(filas, columnas, ...)`. Veamos:

```python
x = np.linspace(0,1,100)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
```

```python
#Dos Ejes
promedio = 500

fig, ax = plt.subplots(2,2, figsize=(8,4))
ax[0,0].set_title(f"$Promedio = {promedio}$")
ax[0,0].set_xlabel("Hola")
ax[0,0].plot(x,y1)
ax[1,1].plot(x,y2)
fig.show()
```

Para este ejemplo se ha establecido el número de filas a 1 y de columnas a 2, además, se ha modificado el tamaño de la figura a (8,4).

```python
x = np.linspace(0,1,100)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
y3 = np.linspace(0,1,100)
y4 = -np.linspace(0,1,100)
```

```python
#Cuatro Ejes
fig, ax = plt.subplots(2,2, figsize=(8,8))
ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
fig.show()
```

Para este ejemplo se ha establecido el número de filas a 2 y de columnas a 2, con el tamaño de la figura a (8,8).    

Ahora, si lo que se quiere es agregar únicamente tres ejes con una configuración particula, entonces deben iniciarcen la figura y los ejes por separado. Para crear la figura se hace uso del método `figure` y para construir eje por eje, se usa `subplot`, de la siguiente forma:

```python
#Tres Ejes
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(221); ax1.plot(x,y1)
ax2 = plt.subplot(222); ax2.plot(x,y2)
ax3 = plt.subplot(212); ax3.plot(x,y3)
fig.show()
```

Para el método `subplot`, la primera entrada indica la candidad de filas, la segunda es el úmero de columnas y la tercera es la posición del eje, comenzando por 1. Adicionalmente, si se desea insertar un eje dentro de otro, entonces se usa el método `add_axes`. Veamos:

```python
x = np.linspace(0,1,100)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
```

```python
#Ejes Superpuestos
fig, ax = plt.subplots(1,1, figsize=(4,4))
#ax2 = ax.twinx()
ax1 = fig.add_axes([0.6,0.65,0.25,0.2])
ax.plot(x,y1)
ax1.plot(x,y2)
fig.show()
```

Para más información sobre la alineación y el espaciado entre ejes visite [Subplots Adjust](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html).

### <font color='green'>**Ejercicio Propuesto 21** </font>

Visualmente, ¿cuál es la diferencia entre una distribución uniforme y una normal?. Genere una figura con dos ejes para ello.

***Ayuda 1:*** Consulte el método `random.uniform`.   
***Ayuda 2:*** Consulte el método `random.normal`.

```python
#Datos
np.random.seed(1320)
μ = 0
σ = 1

x_uniforme = np.random.uniform(-4,4,1000000)
x_normal = np.random.normal(μ,σ, 1000000)
```

```python
min(x_normal), max(x_normal)
```

```python
x = np.linspace(min(x_normal), max(x_normal), 100)
y = (1/(σ*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-μ)/σ)**2)
```

```python
fig, ax = plt.subplots(1,2, figsize=(6,2))
plt.subplots_adjust(wspace=0.3)

ax[0].hist(x_uniforme, bins=100, density=True)
ax[0].set_ylim([0.0,0.5]); ax[0].set_xlim([-5,5])

ax[1].hist(x_normal, bins=100, density=True, label="Datos")
ax[1].set_ylim([0.0,0.5]); ax[1].set_xlim([-5,5])
ax[1].plot(x,y, color="red", label="Gaussiana")
ax[1].legend(loc="upper right")

fig.show()
```
